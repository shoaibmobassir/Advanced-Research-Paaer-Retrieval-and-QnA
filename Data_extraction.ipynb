```python
def XML_DOI(link):

    #defining a header's dictionary to pass through requests
    headers_dict = {"X-ELS-APIKey": "API_KEY", "X-ELS-Insttoken": "Insti_Token", "Accept": "application/xml"}

    #x takes response of the HTTP request, passes link
    x = requests.get(link, headers=headers_dict)

    #Save it as XML file
    with open("doi.xml", 'wb') as f:
        f.write(x.content)
```

```python
# Call the XML_DOI function
for j in range(0, 60):
    xmlfile = XML_DOI(link_list[j])
    # Parse the XML file
    tree = ET.parse('doi.xml')
    root = tree.getroot()

    # Namespace dictionary to handle the XML namespaces
    ns = {
        'prism': 'http://prismstandard.org/namespaces/basic/2.0/',
        'dc': 'http://purl.org/dc/elements/1.1/',
        'atom': 'http://www.w3.org/2005/Atom'
    }

    for entry in root.findall('atom:entry', ns):

        info_dict = {}  # Dictionary to store extracted information

        # Extract data using the appropriate namespaces
        url = entry.find('prism:url', ns)
        info_dict['URL'] = url.text if url is not None else None

        title = entry.find('dc:title', ns)
        info_dict['Title'] = title.text if title is not None else None

        pub_name = entry.find('prism:publicationName', ns)
        info_dict['Pub_Name'] = pub_name.text if pub_name is not None else None

        # Extract cover date with a check for None
        cover_date = entry.find('prism:coverDate', ns)
        info_dict['Cover_Date'] = cover_date.text if cover_date is not None else None

        # Extract DOI with a check for None
        doi = entry.find('prism:doi', ns)
        info_dict['DOI'] = doi.text if doi is not None else None

        # Extract the link with rel="scidir"
        # link = entry.find('.//atom:link[@rel="scidir"]', ns)
        # info_dict['Link'] = link.attrib['href'] if link is not None else None

        # Append the dictionary to the list
        info_list.append(info_dict)

# Extract the information into lists for the DataFrame
DOI = []
Title = []
Pub_name = []
Cover_Date = []
Url = []
# Link = []

for i in range(len(info_list)):
    DOI.append(info_list[i]['DOI'])
    Title.append(info_list[i]['Title'])
    Pub_name.append(info_list[i]['Pub_Name'])
    Cover_Date.append(info_list[i].get('Cover_Date', None))  # Use get() to avoid KeyError
    # Link.append(info_list[i].get('Link', None))  # Use get() to avoid KeyError
    Url.append(info_list[i].get('URL', None))  # Use get() to avoid KeyError

# Create the DataFrame
df = pd.DataFrame()
df['Title'] = Title
df['Pub_name'] = Pub_name
df['DOI'] = DOI
df['Cover_Date'] = Cover_Date
# df['Link'] = Link
df['URL'] = Url

df.head()
```

```python
%%time
list_abstract = []
list_doi= []
list_title = []
list_pubName = []

## ScienceDirect (full-text) document example using DOI
#for i in range(len(data)):
from pprint import pprint
# Access the 'DOI' column as a Series
doi_series = df['DOI']

for i in range(df.shape[0]):
    # Access the doi value using .iloc for position-based indexing
    doi_value = doi_series.iloc[i]

    # Check if doi_value is valid before proceeding
    if doi_value:  # This will skip iterations where doi_value is None
        pprint(doi_value)
        doi_doc = FullDoc(doi=doi_value)
        if doi_doc.read(client):

            #pprint((doi_doc.title))
            abstract = doi_doc._data['coredata']['dc:description']
            title = doi_doc.title
            pubName = doi_doc._data['coredata']['prism:publicationName']
            #pprint(abstract)
            print(i, 'done')
            #dict_abstract['DOI'] = doi[i]
            #dict_abstract['Abstract'] = abstract
            list_abstract.append(abstract)
            list_doi.append(doi_value) # Use doi_value here
            list_title.append(title)
            list_pubName.append(pubName)
            #doi_doc.write()

        else:
            pprint('Operation failed')
            #dict_abstract['DOI'] = doi[i]
            #dict_abstract['Abstract'] = 'NAN'
            list_abstract.append('0')
            list_doi.append('1')
            list_title.append('2')
            list_pubName.append('3')
    else:
        print(f"Skipping iteration {i} due to missing DOI")  # Inform the user about skipped iterations
```
