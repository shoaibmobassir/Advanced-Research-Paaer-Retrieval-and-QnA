```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def search_similar_papers(dfa, keywords, title_weight=0.7, abstract_weight=0.3, top_n=10):
    # Combine Title and Abstract into a new column for TF-IDF vectorization
    dfa_cleaned = dfa.dropna(subset=['Title', 'Abstract'])  # Drop rows with NaN in 'Title' or 'Abstract'
    dfa_cleaned['Title_Abstract'] = dfa_cleaned['Title'] + " " + dfa_cleaned['Abstract']

    # Initialize the TF-IDF vectorizer
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')

    # Fit and transform the combined 'Title_Abstract' column
    tfidf_matrix = tfidf_vectorizer.fit_transform(dfa_cleaned['Title_Abstract'])

    # Vectorize the search keywords using the same vectorizer fitted on the entire dataset
    keyword_vector = tfidf_vectorizer.transform([keywords])

    # Calculate cosine similarity between the keywords and the combined title + abstract
    cosine_sim = cosine_similarity(keyword_vector, tfidf_matrix).flatten()

    # Get the indices of the top 'top_n' most similar papers
    top_indices = cosine_sim.argsort()[-top_n:][::-1]

    # Retrieve the titles and links of the top similar papers
    top_titles = dfa_cleaned.iloc[top_indices]['Title'].tolist()
    top_links = dfa_cleaned.iloc[top_indices]['Link'].tolist()
    top_urls = dfa_cleaned.iloc[top_indices]['Main_URL'].tolist()

    # Return a dictionary with titles and links
    return {"Titles": top_titles, "Links": top_links, "Main_URL":top_urls}

# Example usage:
keywords = "What recent advancements have been made in the photocatalytic efficiency of MoS₂ or other 2D nanomaterials for dye degradation or CO₂ reduction?"
top_results = search_similar_papers(df, keywords)

print("Top 10 most similar papers:")
for i, (title, link) in enumerate(zip(top_results['Titles'], top_results['Links']), 1):
    print(f"{i}. {title} - {link}")
```

```python
import os

from langchain_community.document_loaders import UnstructuredPDFLoader, DirectoryLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_groq import ChatGroq
from langchain.chains import RetrievalQA
```

```python
import os
# Create directories for data if they don't exist
os.makedirs("data", exist_ok=True)
os.makedirs("data/uploaded_pdfs", exist_ok=True)
```

```python
loader = DirectoryLoader("data/uploaded_pdfs/", glob="./*.pdf", loader_cls=UnstructuredPDFLoader)
documents = loader.load()
```

```python
from google.colab import userdata
GROQ_API_KEY=userdata.get('GROQ_API_KEY')
```

```python
text_splitter = CharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=500
)


# If getting warnings like('WARNING:langchain_text_splitters.base:Created a chunk of size 2045, which is longer than the specified 2000')
# Then you may use either of both. Just "Uncomment either of two"
# Use
# text_splitter = RecursiveCharacterTextSplitter(chunk_size=1900, chunk_overlap=150)
# or
# from langchain.text_splitter import TokenTextSplitter
# text_splitter = TokenTextSplitter(chunk_size=2000, chunk_overlap=100)

text_chunks = text_splitter.split_documents(documents)
```

```python
persist_directory = "data/doc_db"
```

```python
embedding = HuggingFaceEmbeddings()
```

```python
vectorstore = Chroma.from_documents(
    documents=text_chunks,
    embedding=embedding,
    persist_directory=persist_directory
)
```

```python
retriever = vectorstore.as_retriever()
```

```python
os.environ["GROQ_API_KEY"] = GROQ_API_KEY
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "Explain LSPR theory in non-metal plasmonic."
response = qa_chain.invoke({"query":query})
```
